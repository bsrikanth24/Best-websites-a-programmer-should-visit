```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import split, trim

# Initialize Spark session
spark = SparkSession.builder.appName("SplitFullName").getOrCreate()

# Sample data
data = [
    ("Joe Root",),
    ("Virat Kohli",),
    ("Kane Williamson",),
    ("AB de Villers",),
    ("Christopher Nolan",)
]

# Create DataFrame
df = spark.createDataFrame(data, ["FullName"])

# Split FullName into firstName and lastName
df_split = df.withColumn("firstName", trim(split(df["FullName"], " ")[0])) \
             .withColumn("lastName", trim(split(df["FullName"], " ")[-1]))

df_split.show(truncate=False)
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import split, trim, col, element_at, size

# Initialize Spark session
spark = SparkSession.builder.appName("SplitFullName").getOrCreate()

# Sample data
data = [
    ("Joe Root",),
    ("Virat Kohli",),
    ("Kane Williamson",),
    ("AB de Villers",),
    ("Christopher Nolan",)
]

# Create DataFrame
df = spark.createDataFrame(data, ["FullName"])

# Split FullName into array of words
split_col = split(col("FullName"), " ")

# Extract firstName (first element)
first_name = trim(split_col.getItem(0))

# Extract lastName (last element)
last_name = trim(element_at(split_col, size(split_col)))

# Add columns to DataFrame
df_split = df.withColumn("firstName", first_name) \
             .withColumn("lastName", last_name)

df_split.show(truncate=False)
```

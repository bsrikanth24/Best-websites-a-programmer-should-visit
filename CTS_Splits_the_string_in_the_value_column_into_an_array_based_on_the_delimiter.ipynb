```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import split, col

# Initialize Spark session
spark = SparkSession.builder.appName("SplitColumnExample").getOrCreate()

# Create input DataFrame
data = [(1, "a,b,c"), (2, "d,e,f")]
columns = ["col1", "value"]
df = spark.createDataFrame(data, columns)

# Show input DataFrame
df.show()
```

```python
# Split the 'value' column into an array
df_split = df.withColumn("value_split", split(col("value"), ","))

# Select and create new columns for each element in the array
df_transformed = df_split.select(
    col("col1"),
    col("value_split")[0].alias("col2"),
    col("value_split")[1].alias("col3"),
    col("value_split")[2].alias("col4")
)

# Show the transformed DataFrame
df_transformed.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import split, col

# Initialize Spark session
spark = SparkSession.builder.appName("SplitColumnExample").getOrCreate()

# Create input DataFrame
data = [(1, "a,b,c"), (2, "d,e,f")]
columns = ["col1", "value"]
df = spark.createDataFrame(data, columns)

# Transform the DataFrame directly using withColumn
df_transformed = (
    df
    .withColumn("col2", split(col("value"), ",")[0])
    .withColumn("col3", split(col("value"), ",")[1])
    .withColumn("col4", split(col("value"), ",")[2])
    .drop("value")  # Drop the original 'value' column if not needed
)

# Show the transformed DataFrame
df_transformed.show()
```

```python
from pyspark.sql import SparkSession

# Initialize SparkSession if not already active
spark = SparkSession.builder.appName("DataFrameOperations").getOrCreate()

# Create a sample DataFrame 'df' with the 'Customerkeycode' column
data = [("PLANNING1:DPG1:APG1",),
        ("PLANNING2:DPG2:APG2",),
        ("PLANNING3:DPG3:APG3",),
        ("B01:B14:110083",)] # Adding user-provided example
columns = ["Customerkeycode"]
df = spark.createDataFrame(data, columns)

df.show(truncate=False)
```

```python
import pyspark.sql.functions as F

cols = ['PlanningCustomerSuperGroupCode', 'DPGCode', 'APGCode']
split_col = F.split('Customerkeycode', ':')
arr_cols = [split_col[i].alias(cols[i]) for i in range(3)]

df = df.select(*arr_cols)
df.show(truncate=False)
```

```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F

# Re-initialize SparkSession if not already active
spark = SparkSession.builder.appName("StringsplitOperations").getOrCreate()

# Re-create df1 to its original state to ensure 'Fullname' is present
data1 = [("Srikanth Bommakanti",),
        ("Jayashree Earamyna",),
        ("Sneha Palle",),
        ("Bhavan Reddy",)]

column = ["Fullname"]
df1 = spark.createDataFrame(data1, column)

cols = ['First Name', 'Last Name']
split_col = F.split('Fullname', ' ')

# Now, with df1 correctly initialized, perform the selection including original and new split columns
arr_cols = [
    F.col('Fullname')
]
arr_cols.extend([split_col[i].alias(cols[i]) for i in range(2)])

df1 = df1.select(*arr_cols)
df1.show(truncate=False)
```

```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F

# Re-initialize SparkSession if not already active (ensure spark object exists)
spark = SparkSession.builder.appName("DataFrameOperations").getOrCreate()

# Re-create df2 to its original state to ensure 'col1' and 'my_str_col' are present
data = [
    (18, "856-yygrm"),
    (201, "777-psgdg")
]
columns = ["col1", "my_str_col"]
df2 = spark.createDataFrame(data, columns)

cols2 = ['_col3','_col4']
split_col = F.split('my_str_col', '-')

# Now, with df2 correctly initialized, perform the selection including original and new split columns
arr_cols = [
    F.col('col1'),
    F.col('my_str_col')
]
arr_cols.extend([split_col[i].alias(cols2[i]) for i in range(2)])

df2 = df2.select(*arr_cols)
df2.show(truncate=False)
```

```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F

spark = SparkSession.builder.appName("StringsplitOperations").getOrCreate()

data1 = [
    ("Srikanth Bommakanti",),
    ("Jayashree Earamyna",),
    ("Sneha Palle",),
    ("Bhavan Reddy",),
]

df1 = spark.createDataFrame(data1, ["Fullname"])

cols = ['First Name', 'Last Name']
split_col = F.split('Fullname', ' ')

# No .extend() — just +
arr_cols = [F.col('Fullname')] + [split_col[i].alias(cols[i]) for i in range(2)]

df1 = df1.select(*arr_cols)
df1.show(truncate=False)
```
